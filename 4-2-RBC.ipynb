```python
import numpy as np
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
import sympy as sp
sp.init_printing()

from sympy import Q
from sympy.assumptions.assume import global_assumptions

import japanize_matplotlib
```

```python
Ct, Lt, Kt, Yt, At = sp.symbols("C_{t}, L_{t}, K_{t}, Y_{t}, A_{t}")
Ca, Ka, Aa = sp.symbols("C_{t+1}, K_{t+1}, A_{t+1}")
wt, rt, ra, ea = sp.symbols("w_{t}, r_{t}, r_{t+1}, \\epsilon_{t+1}")
beta, mu, gamma, alpha, delta, rho = \
    sp.symbols("beta, mu, gamma, alpha, delta, rho")

global_assumptions.add(Q.is_true(alpha > 0))
global_assumptions.add(Q.is_true(alpha < 0))
global_assumptions.add(Q.is_true(gamma > 0))
```

```python
ea
```

```python
EQM = sp.Matrix([
    wt / Ct - (gamma + 1) * mu * Lt**gamma,
    Ca / Ct - beta * (ra - delta + 1),
    Yt - At * Kt**alpha * Lt**(1 - alpha),
    wt - (1 - alpha) * Yt / Lt,
    Ka - Yt - (1 - delta) * Kt + Ct,
    rt - alpha * Yt / Kt,
    sp.log(Aa) - rho * sp.log(At) - ea
])

EQM
```

```python
ss_symb = EQM.subs({
    Ca: Ct,
    Ka: Kt,
    Aa: At,
    ra: rt,
    ea: 0
})
ss_symb
```

```python
sp.solve(ss_symb)
```
